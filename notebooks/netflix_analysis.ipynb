```python
import pandas as pd

df = pd.read_csv("../data/netflix_titles.csv")
```

```python
df['director'] = df['director'].fillna("No Director")
df['cast'] = df['cast'].fillna("No Cast")
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Not Rated")
df['date_added'] = df['date_added'].fillna("Unknown")
df['duration'] = df['duration'].fillna("Not Available")
```

```python
df['listed_in'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Genres on Netflix")
plt.xticks(rotation=45)
plt.savefig("../images/genres.png")
plt.show()
```

```python
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Movies vs TV Shows")
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot()
plt.title("Content Growth Over Years")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top Countries Producing Content")
plt.xticks(rotation=45)
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top Countries Producing Content")
plt.xticks(rotation=45)
plt.show()
```

```python
df['rating'].value_counts().plot(kind='bar')
plt.title("Ratings Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
movies = df[df['type'] == 'Movie']
movies['duration'].value_counts().head(10).plot(kind='bar')
plt.title("Movie Duration Distribution")
plt.xticks(rotation=45)
plt.show()
```
